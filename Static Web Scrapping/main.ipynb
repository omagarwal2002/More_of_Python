```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait, Select
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException

# Establish chrome driver and go to report site URL
url = "https://cricbuzz.com/cricket-series/3472/indian-premier-league-2021/matches/"
driver = webdriver.Chrome(r"C:\Users\Dell\Downloads\chromedriver.exe")
driver.get(url)
```

```python
match = []
match.clear()
file_name = "D:\\vs code\\MLProject_Sem4\\Data.txt"
file_read = open(file_name, "r")
text = "vs"
lines = file_read.readlines()
idx = 0

for line in lines:

        if text in line:
            match.append(line)
            idx += 1

file_read.close()

if len(match) == 0:
        print("\n\"" + text + "\" is not found in \"" + file_name + "\"!")
else:

        lineLen = len(match)
        print("\n**** Lines containing \"" + text + "\" ****\n")
        for i in range(lineLen):
            print(end=match[i])
        print()
```

```python
with open('D:\\vs code\\MLProject_Sem4\\Matches.txt',"w",encoding="utf-8") as f:
    f.writelines(match)
```

```python
dates = []
dates.clear()
file_name = "D:\\vs code\\MLProject_Sem4\\Dates.txt"
file_read = open(file_name, "r")
text = ","
lines = file_read.readlines()
idx = 0

to_repeat = ["Apr 18, Sun\n","Apr 21, Wed\n","Apr 25, Sun\n","Apr 29, Thu\n","May 02, Sun\n","Sep 25, Sat\n","Sep 26, Sun\n","Sep 28, Tue\n","Oct 02, Sat\n","Oct 03, Sun\n","Oct 07, Thu\n","Oct 08, Fri\n"]
for line in lines:

        if text in line:
            dates.insert(idx,line)
            idx += 1
for val in to_repeat:
        if val in dates:
                dates.insert(dates.index(val)+1,val)
                idx += 1
       
           
file_read.close()

if len(dates) == 0:
        print("\n\"" + text + "\" is not found in \"" + file_name + "\"!")
else:

        lineLen = len(dates)
        print("\n**** Lines containing \"" + text + "\" ****\n")
        for i in range(lineLen):
            print(end=dates[i])
        print()
```

```python
with open('D:\\vs code\\MLProject_Sem4\\Fixtures.txt',"w",encoding="utf-8") as f:
    f.writelines(dates)
```

```python
times = []
times.clear()
file_name = "D:\\vs code\\MLProject_Sem4\\Data.txt"
file_read = open(file_name, "r")
text = "7:30 PM"
text2= "3:30 PM"
text3= "07:30 PM"
text4= "03:30 PM"
lines = file_read.readlines()
idx = 0

for line in lines:

        if text in line or text2 in line:
            times.append(line)
            idx += 1
for line in lines:
    
        if text3 in line or text4 in line:
            times.remove(line)


file_read.close()

if len(times) == 0:
        print("\n\"" + text + "\" is not found in \"" + file_name + "\"!")
else:

        lineLen = len(times)
        print("\n**** Lines containing \"" + text + "\" ****\n")
        for i in range(lineLen):
            print(end=times[i])
        print()
```

```python
with open('D:\\vs code\\MLProject_Sem4\\Timings.txt',"w",encoding="utf-8") as f:
    f.writelines(times)
```
